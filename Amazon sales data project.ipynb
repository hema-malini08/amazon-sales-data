Data Cleaning

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
for col in df.select_dtypes(include='number').columns:
    df[col].fillna(df[col].median(), inplace=True)

for col in df.select_dtypes(exclude='number').columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['total_reviews'] = df['total_reviews'].astype(int)
```

```python
df['purchased_last_month'] = df['purchased_last_month'].astype(int)
```

```python
df['delivery_date'] = pd.to_datetime(df['delivery_date'], errors = 'coerce')
df['data_collected_at'] = pd.to_datetime(df['data_collected_at'], errors = 'coerce')
```

```python
df.drop(['sustainability_tags'],axis = 1, inplace = True)
```

```python
df.info()
```

Data Analysis

```python
df.groupby('is_sponsored')['purchased_last_month'].mean()
```

```python
df.groupby('product_category')['purchased_last_month'].sum().sort_values(ascending=False)
```

Key Insights
1. Sponsored vs Organic Products

Sponsored products show significantly higher average purchases than organic products.
Indicates that paid promotions improve visibility and sales performance.

2. Category Performance
Power & Batteries accounts for 26.17M purchases, representing ~60–65% of total demand
Significantly outperforms all other categories.
